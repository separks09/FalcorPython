```python
#import dependencies
import pandas as pd
import json
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
#read in json files, set up dataframes
ufodict=[]
ufo_reports = open("nuforc_reports.json")
for line in ufo_reports:
    ufodict.append(json.loads(line))
    
bfrodict=[]
bfro_reports = open("bfro_reports.json")
for line in bfro_reports:
    bfrodict.append(json.loads(line))
    
ufo = pd.DataFrame.from_dict(ufodict)
bigfoot = pd.DataFrame.from_dict(bfrodict)
```

```python
#Reports by State, % by state?
bigfootstate = bigfoot.dropna(subset=["STATE"])
bigfootstate = bigfootstate.groupby("STATE")
footcount=bigfootstate["STATE"].count()
footcount=pd.DataFrame(footcount)
footcount=footcount.sort_values("STATE", ascending=False)
footcount.head(10)
```

```python
ufostate = ufo.dropna(subset=["state"])
ufostate = ufostate.groupby("state")
ufocount=ufostate["state"].count()
ufocount=pd.DataFrame(ufocount)
ufocount=ufocount.sort_values("state", ascending=False)
ufocount.head(10)
```

```python
#Chart the Bigfoot Data
plt.title("Top Ten Bigfoot States")
plt.pie(footcount.head(10), labels=["Washington","California","Florida","Ohio", "Oregon", "Illinois", "Texas", "Michigan", "Missouri", "Georgia"], autopct="%.1f%%")
plt.axis("equal")
plt.show()

```

```python
#Chart the UFO Data
plt.title("Top Ten UFO States")
plt.pie(ufocount.head(10), labels=["California","Florida","Washington","Texas", "New York", "Arizona", "Pennsylvania", "Illinois", "Ohio", "Michigan"], autopct="%.1f%%")
plt.axis("equal")
plt.show()
```
